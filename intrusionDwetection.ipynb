```python
x = df.iloc[::, 0:85:]
y = df.iloc[::, -1]
```

```python
from imblearn.under_sampling import RandomUnderSampler
ros = RandomUnderSampler()
x_over, y_over = ros.fit_resample(x,y)
print("Imbalanced datapoint :", y.value_counts())
print()
print("Balanced datapoint :", y_over.value_counts())
```

```python
x_over.to_csv('underSampledBalancedDataSet_IntrusionDetectionFeature.csv')
```

```python
y_over.to_csv('underSampledBalancedDataSet_IntrusionDetectionTarget.csv')
```

```python
# Split the data into training and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_over, y_over, test_size=0.2, random_state=101,
                                                   stratify= y_over)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=21)
knn.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
y_pred_train_knn = knn.predict(x_train)
y_pred_test_knn = knn.predict(x_test)
```

```python
print(accuracy_score(y_train, y_pred_train_knn))
print()
print(accuracy_score(y_test, y_pred_test_knn))
```
